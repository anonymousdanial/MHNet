```python
import matplotlib.pyplot as plt
import re
```

```python
def load_losses_from_log(filepath):
    epochs = []
    losses = []

    # Regex to capture: Epoch 5: 15152.178678
    pattern = re.compile(r"Epoch\s+(\d+):\s+([\d\.]+)")

    with open(filepath, "r") as file:
        for line in file:
            match = pattern.search(line)
            if match:
                epoch = int(match.group(1))
                loss = float(match.group(2))
                epochs.append(epoch)
                losses.append(loss)

    return epochs, losses

def plot_loss(epochs, losses):
    plt.figure()
    plt.plot(epochs, losses)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss per Epoch")
    plt.grid(True)
    plt.show()
```

```python
log_file = "LPE.txt"  # Change to your actual .txt file name
epochs, losses = load_losses_from_log(log_file)
plot_loss(epochs, losses)
```
